# In-N-Out Burger Problem 

The In-N-Out Burger Problem can be simply stated thus:

> You frequently go to In-N-Out Burger. Each time you go, you are given a random number between 1 and 99. How many visits $v$ will it take to accumulate all 98 numbers? 

```python
%matplotlib inline
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
v = range( 0, 1000 )
x = []
y = []
a = []

for visit_number in v:
    x.append( visit_number )
    a.append( np.random.choice( range( 1, 99 ) ) )
    y.append( len( set( a ) ) )
```

```python
plt.figure()
plt.scatter( x, y, color='k', marker='.' )
plt.xlabel( 'Number of visits' )
plt.ylabel( 'Unique numbers accumulated' )
```

**Answer**. Emperically, it looks like after about 300 visits, you're likely to have accumulated all 98 numbers.
